### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,11)
y = x**2
plt.plot(x, y)
plt.title("Line Plot of y = X^2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

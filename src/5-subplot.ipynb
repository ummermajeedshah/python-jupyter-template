### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = np.sin(x)
bar_categories = ['A', 'B', 'C', 'D']
bar_values = [3, 7, 5, 9]
scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)
pie_labels = ['Apple', 'Banana', 'Cherry', 'Date']
pie_sizes = [20, 30, 35, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Line Chart")

# Bar chart
axs[0, 1].bar(bar_categories, bar_values, color='green')
axs[0, 1].set_title("Bar Chart")

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='red')
axs[1, 0].set_title("Scatter Plot")

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'pink', 'lightgreen'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()

```
